```python
# import all libraries
import pandas as pd
import numpy as np
```

```python
# read csv files
data_dir = '../data/'
df_defendant_docket_complete = pd.read_csv(data_dir+'defendant_docket_complete.csv')
df_docket_details = pd.read_csv(data_dir+'defendant_docket_details.csv')
df_docket_ids = pd.read_csv(data_dir+'defendant_docket_ids.csv')
df_offenses_disposition = pd.read_csv(data_dir+'offenses_dispositions_v3_cleaned_peroid.csv')
```

### Merge on offenses_dispositions and defendant_docket_complete

```python
# https://stackoverflow.com/questions/54221673/is-there-a-way-to-export-pandas-dataframe-info-df-info-into-an-excel-file
def get_dataframe_info(df):
    """
    input
       df -> DataFrame
    output
       df_null_counts -> DataFrame Info (sorted)
    """

    df_types = pd.DataFrame(df.dtypes)
    df_nulls = df.count()
    
    df_null_count = pd.concat([df_types, df_nulls], axis=1)
    df_null_count = df_null_count.reset_index()
    
    # Reassign column names
    col_names = ["variables", "dtypes", "non_null_counts"]
    df_null_count.columns = col_names
    
    
    return df_null_count
get_dataframe_info(df_original)
```

```python
with pd.ExcelWriter('offenses_dispositions_v4_variable_description.xlsx') as writer:
    df_original.head(10).to_excel(writer, sheet_name='preview')
    get_dataframe_info(df_original).to_excel(writer, sheet_name='variables',index=False)
```
